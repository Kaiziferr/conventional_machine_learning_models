# **Funciones**
---

```python
def normalize_word(word):
  """Normalization of names of features"""
  word = word.replace(' ', '_')
  find_guion = word.find('_')
  list_word = []
  if find_guion:
    list_word = [w for w in word.split('_') if w != '']
  else:
    list_word = word
  word = list(map(lambda x: x.lower(), list_word))
  word = [normalize('NFKD', c).encode('ASCII', 'ignore').decode() for c in word]
  word = "_".join(word)
  return word


def normalize_name_columns(columns):
  """Normalization of columns"""
  columns = list(map(lambda x: normalize_word(x), columns))
  return columns


def tree_plot(model, X_columns, y_columns):
  fig, ax = plt.subplots(figsize=(12, 5))
  print(f"Tree Deep: {model.get_depth()}")
  print(f"Number of terminal nodes: {model.get_n_leaves()}")

  plot = plot_tree(
    decision_tree = model,
    feature_names = X_columns,
    class_names   = y_columns,
    filled        = True,
    impurity      = False,
    fontsize      = 5,
    precision     = 2,
    ax            = ax
)

def importances_feature(model, X_columns):
  """important predictors table"""
  importancia_predictores = pd.DataFrame(
      {'predictor': X_columns,
      'importance': model.feature_importances_*100}
      )

  return importancia_predictores.sort_values('importance', ascending=False)
```

```python
url_miner = 'https://drive.google.com/file/d/10_YGM_dOKzzLsLNkfB_bN5YeLgDHcrUZ/view?usp=sharing'
url_miner = 'https://drive.google.com/uc?id=' + url_miner.split('/')[-2]
data_miner = pd.read_csv(url_miner, dtype='str')
```

- Naming assignment

- Asignación de nombres

```python
data_miner.columns = [
    'producing_municipality',
    'department',
    'natural_resource',
    'project_name',
    'production_year',
    'quarter',
    'unit_measure',
    'compensation',
    'amount',
    'production quantity'
]
```

- Assignment the real type

- Asignación del tipo de dato real

```python
data_miner[[
    'amount',
    'production_year'
]] = data_miner[[
    'amount',
    'production_year'
]].astype('float64')
```

- A 5-year data period will be used for model training.

- Se tomará como periodo de datos para el entrenamiento de modelos 5 años

```python
data_miner = data_miner[
    (data_miner['production_year'] >= 2018)
    &
    (data_miner['production_year'] <= 2022)
]
```

- The value of the quarter variable's string is modified to extract the numbers, allowing for easier manipulation of the variable.

- El valor de la cadena de la variable de trimestre se modifica para extraer los números, lo que permite una manipulación más sencilla de la variable.

```python
data_miner['quarter'] = data_miner[
    'quarter'].apply(
        lambda x: x.split(' ')[1]
    )

data_miner.head(3)
```

- The department and state columns are concatenated to create a column that can uniquely identify the records, since different departments may have cities with the same name.

- Las columnas de departamento y estado se concatenan para crear una columna que pueda identificar de forma única los registros, ya que diferentes departamentos pueden tener ciudades con el mismo nombre.

```python
data_miner['key'] = data_miner.producing_municipality.str.cat(
    data_miner.department, sep='-')
```

```python
data_miner.head()
```

- Grouping of features transactional
- Agrupación de características transaccionales

```python
function_dictionary = {
    'amount': ['sum', 'count']
}

filter_feature = [
    'key',
    'production_year',
    'quarter',
    'compensation'
]

new_val_col = [
    'sum_compensation',
    'count_compensation'
]

data_group = data_miner.groupby(
    filter_feature
).aggregate(function_dictionary)

data_group.columns = new_val_col
data_group = data_group.reset_index()

data_group.head()
```

```python
function_dictionary = {
    "sum_compensation": ['sum'],
    "count_compensation": ['sum'],
    'quarter': ['count'],
    'production_year': ['nunique']
}

filter_feature = [
    'key',
    'compensation']



data_group2 = data_group.groupby(
    filter_feature).aggregate(function_dictionary)

new_val_col = [
    'total_compensation',
    'count_transactions',
    'number_active_quarters',
    'year_production_active',
]


data_group2.columns = new_val_col
data_group2 = data_group2.reset_index()

data_group2.head()
```

# **Split**
---

- The dataset is split into training and validation data.
- El conjunto de datos se divide en datos de entrenamiento y validación.

```python
data_numeric = data_group2.select_dtypes(include=['float64', 'int64']).columns.to_list()
data_cat = data_group2.select_dtypes(include=['object']).columns.to_list()
```

```python
X = data_group2[data_numeric[1:]+data_cat[1:]]
y = data_group2[data_numeric[0]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state = random_seed
)
```

#**Preprocessing and Model**
---

- I will first use the "OneHotEncoder" method without eliminating a variable.
- Primero utilizaré el método "OneHotEncoder" sin eliminar una variable.

```python
compose_column = ColumnTransformer(
    [
        ('OneHot', OneHotEncoder(), data_cat[1:])
    ],
    remainder='passthrough'
)

X_train_compose = compose_column.fit_transform(X_train)
X_test_compose = compose_column.transform(X_test)

encoded_cat = compose_column.named_transformers_['OneHot'].get_feature_names_out(
    data_cat[1:]).tolist()

col = encoded_cat + data_numeric[1:]
col
```

```python
X_train_method =pd.DataFrame(X_train_compose, columns = col)
X_test_method =pd.DataFrame(X_test_compose, columns = col)
```

- A model with a depth of 3 will be used as a baseline.

- Se utilizará un modelo con una profundidad de 3 como línea base.

```python
model_base = DecisionTreeRegressor(
    max_depth = 3,
    random_state = random_seed,
)

model_base.get_params()
```

```python
model_base = DecisionTreeRegressor(
    max_depth = 3,
    random_state = random_seed,
)
params_base_master =  model_base.get_params()
model_base.fit(X_train_method, y_train)
importances_feature(model_base, X_train_method.columns)
```

- The preprocessing of this dataset has the peculiarity that there is a single important feature, while the rest have no effect on the tree split.

- El preprocesamiento de este conjunto de datos tiene la particularidad de que hay una única característica importante, mientras que el resto no tiene efecto en la división del árbol.

- One possible cause could be the limited variation of the other variables compared to the most important one (number of transactions).

- Una posible causa podría ser la variación limitada de las otras variables en comparación con la más importante (número de transacciones).
